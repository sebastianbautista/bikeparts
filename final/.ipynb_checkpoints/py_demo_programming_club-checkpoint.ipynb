# Python demo - Programming Club

In this notebook, I use pandas, matplotlib, and seaborn to aggregate trip-level Capital Bikeshare data to the daily level and demonstrate basic plotting.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
sns.set_palette('colorblind')
```

```python
# Read in data - 2017 Capital Bikeshare data
# Each row is a trip
df = pd.read_csv(r'~/CaBi/Output/CABI_Daily_Trips_20180315_191537_2017-q1_2017-q2_2017-q3_2017-q4.csv', 
                 dtype={'Start station number': object,
                       'End station number': object,
                       'start_region_code': object,
                       'end_region_code': object,
                       'region_start_end': object})
df.head()
```

```python
# We want to roll this data up to the daily level and generate counts of bike rides by day and by membership status
#total_count = []
def daily_total(df):
    daily_df = df.groupby('start_date').size()
    #total_count.append(daily_df)
    return daily_df

#member_count = []
def member_total(df):
    member_df = df.groupby(['start_date', 'Member type']).size().unstack(level=-1).reset_index()
    #member_count.append(daily_df)
    return member_df

# We're only looking at a single year of data here, but defining functions is useful for dealing with larger amounts of data
total_count = daily_total(df)
member_count = member_total(df)
member_count.head()
```

```python
# daily_total() returns a Series - no column name and start_date is the index
total_count.head()
```

```python
# Pull start_date out of index and rename count
total_count = total_count.reset_index().rename(columns={0: 'Total'})
total_count.head()
```

```python
# Create final dataframe 
final = member_count.merge(total_count, how='outer', on='start_date')
final.head()
```

```python
# Datetime variables
final['date'] = pd.to_datetime(final['start_date'])
final['year'] = final['date'].dt.year 
final['month'] = final['date'].dt.month
final['weekday'] = final['date'].dt.weekday
final['weekday_name'] = final['date'].dt.weekday_name
final['quarter'] = final['date'].dt.quarter

final.head()
```

```python
# Plot
ax1 = plt.subplot(121)
sns.boxplot(x=final['quarter'], y=final['Total'])

ax2 = plt.subplot(122, sharey=ax1)
sns.violinplot(x=final['quarter'], y=final['Total'])
plt.setp(ax2.get_yticklabels(), visible=False)

plt.show()
```

```python
sns.swarmplot(x=final['month'], y=final['Total'])
plt.show()
```

```python
sns.swarmplot(x=final['weekday'], y=final['Total'], hue=final['quarter'])
plt.show()
```

```python
sns.pairplot(final, vars=['Casual', 'Member', 'Total'], hue='quarter')
plt.show()
```

```python
sns.pairplot(final, vars=['Casual', 'Member', 'Total'], kind='reg', diag_kind='kde')
plt.show()
```

```python
sns.pairplot(final, hue='quarter')
plt.show()
```
